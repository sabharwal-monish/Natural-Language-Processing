```python
import spacy
```

```python
from google.colab import files
uploaded = files.upload()

with open('news_story.txt', 'r') as f:
    news_text = f.read()
```

```python
news_text[:500]
```

```python
nlp = spacy.load('en_core_web_sm')

numeral_tokens = []
noun_tokens = []

doc = nlp(news_text)
for token in doc:
    if token.pos_ == "NOUN":
        noun_tokens.append(token)
    elif token.pos_ == 'NUM':
        numeral_tokens.append(token)

print(numeral_tokens)
print(noun_tokens)
```
