```python
import spacy

# Load the pre-trained large English model
nlp = spacy.load("en_core_web_lg")

# Create a Doc object by processing a text
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

# Access word vectors for each token
for token in doc:
    print(f"Token: {token.text} | Vector: {token.vector[:5]}...")
```

```python
word1 = nlp("apple")
word2 = nlp("banana")
similarity = word1.similarity(word2)
print(f"Similarity between 'apple' and 'banana': {similarity:.4f}")
```
